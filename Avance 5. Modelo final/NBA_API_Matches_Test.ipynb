```python
for abreviacion, id in teams_ids.items():
  print(f"\nEquipo: {abreviacion}")

  if id in teams_processed:
    print(f"✅ Información ya obtenida\n")
    continue

  print(f"❗ Obteniendo información...\n")
  for season in SEASONS_LIST:
    print(f"Temporada {season}")

    team_log = TeamGameLogs(team_id_nullable=id,
                            season_nullable=season,
                            date_from_nullable=date_from)

    # Son milisegundos
    for i in tqdm(range(10), desc="Time out"):
      time.sleep(0.1)

    print(team_log)


    df_team = team_log.get_data_frames()[0]

    teams_logs_dfs.append(df_team)

  teams_processed.add(id)
  print(f"\nNo de equipos acumulados: {len(teams_logs_dfs)}\n")
  # Son milisegundos
  for i in tqdm(range(50), desc="Time out"):
    time.sleep(0.1)
```

```python
teams_df = pd.concat(teams_logs_dfs, ignore_index=True)
teams_df['GAME_DATE'] = pd.to_datetime(teams_df['GAME_DATE'])
display(teams_df.shape)
display(teams_df)
```

```python
teams_df.to_csv(f'teams_matches_nba_from_{date_from}_to_{date_to}.csv')
```
